```python
import pandas as pd
import numpy as np
import dask.dataframe as dd
import datetime as dt
```

# Leyendo Matrices OD del d√≠a 16 al 20 de marzo del 2020

```python
tabla=dd.read_csv(f'input/2012abril_16_20.MatrizODviajes_comunas_mediahora.csv', sep='|').compute()
tabla.head()
```

```python
tabla=tabla.replace({',': '.'}, regex=True)
tabla['mediahora']  = '2012-04-16 ' + tabla['mediahora']
tabla['hora_subida']=pd.to_datetime(tabla['mediahora']) #creo la columna hora de subida en formato datatime
tabla.set_index('hora_subida', inplace=True) #fijo la columna como √≠ndice
tabla.head()
```

```python
tabla=tabla.between_time('06:00', '10:00', include_end=False).reset_index() #filtro los datos en el tiempo que me interesa
tabla
```

```python
tabla=tabla.astype({'viaje_laboral_promedio':'float64','viajes_1_etapa': 'float64', 'viajes_2_etapas':'float64',
                    'viajes_3_etapas':'float64', 'viajes_4_o_mas_etapas':'float64'}) #convierto a float las columnas que me importan
```

```python
tabla.head()
```

```python
tabla['viajes']=tabla[['viaje_laboral_promedio']]
tabla['day']='16' # es un promedio de los dias de semana les asigno los viajes al d√≠a 16 por consistencia de datos del 2020
tabla.head()
```

# DATOS FINALES

```python
viajes=tabla[['day','comunasubida', 'comunabajada', 'viajes']].sort_values(by=['day','comunasubida', 'comunabajada'])
```
